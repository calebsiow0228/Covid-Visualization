```python
#Looking for the percent of missing data
for col in deaths.columns:
    pct_missing = np.mean(deaths[col].isnull()) * 100
    print(col, pct_missing)
```

```python
#Dropping data that have more than 60% missing data
deaths.drop("weekly_hosp_admissions_per_million", axis = 1, inplace = True)
```

```python
#Dropping weekly_hosp_admission 
deaths.drop("weekly_hosp_admissions", axis = 1,inplace = True)
```

```python
#Dropping weekly icu admission per million
deaths.drop("weekly_icu_admissions_per_million", axis = 1, inplace = True)
```

```python
#Dropping weekly icu admission
deaths.drop("weekly_icu_admissions", axis = 1, inplace = True)
```

```python
#Dropping hosp_patients_per_million
deaths.drop("hosp_patients_per_million", axis = 1, inplace = True)
```

```python
#Dropping hosp_patients
deaths.drop("hosp_patients", axis = 1, inplace = True)
```

```python
#dropping icu_patients_per_million
deaths.drop("icu_patients_per_million", axis = 1, inplace = True)

#Dropping icu_patients
deaths.drop("icu_patients", axis = 1, inplace = True)
```

```python
#categorizing data into list("Object")
object_columns = deaths.columns[(deaths.dtypes == 'object').values].tolist()
object_columns
```

```python
numerical_columns = deaths.columns[(deaths.dtypes != 'object').values].tolist()
numerical_columns
```

```python
#Finding the unique values of each object category
for col in object_columns:
    print(col, '\n')
    print(deaths[col].value_counts())
    print('---------------------------------------------')
```

```python
#Display unique values of numerical columns
for col in numerical_columns:
    print(col, '\n')
    print(deaths[col].value_counts())
    print('------------------------------------------------')
```

```python
#Covert population to population per million
deaths['populations'] = round(deaths['population']/1000000, 3)
deaths.rename(columns ={'population': 'ppm'}, inplace = True)
```

```python
deaths.head()
```

```python
#USA Covid Deaths Data
USA = deaths[deaths['location'].isin(['United States'])].copy()
USA.head()
```

```python
#Displaying united states covid cases over time


dates = USA['date'].unique().tolist()
dates.sort()
plt.figure(figsize=(16,10))
plt.plot(dates, USA['total_cases'])
plt.title('# of Covid-19 Cases in the US Over Time')
plt.xlabel("From 1/22/2020 to 7/8/2021", size = 30)
plt.ylabel('# of Cases', size = 30)
plt.legend(['USA Covid-19 cases over time'], prop = {'size' : 20})
plt.xticks(size = 20)
plt.yticks(size = 20)
plt.show()
```
