```python
#importing modules
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
```

```python
#importing datasets
covid_dataset = pd.read_csv("covid_confirmed.csv")
covid_dataset.head(10)
```

```python
#Checkings the shape of the dataframe
covid_dataset.shape
```

```python
#Delete the useless columns
covid_dataset.drop(["Lat", "Long"], axis = 1, inplace = True)
```

```python
#check updated dataframe
covid_dataset.head()
```

```python
#Aggregating the rows by the country
covid_agg = covid_dataset.groupby("Country/Region").sum()
```

```python
#Displaying aggregated covid dataset
covid_agg.head()
```

```python
#Check shape of aggregated Dataset
covid_agg.shape
```

```python
#Visualizing Data: Comparing Covid in China to Covid in US
covid_agg.loc["China"].plot()
covid_agg.loc["US"].plot()
covid_agg.loc["Italy"].plot()
plt.legend()
```

```python
#Calculating a good measure
```

```python
#First three days of covid in china
covid_agg.loc["China"][:3].plot() #from 1/22/20 to 1/23/20 new confirmed cases grew from 550 to 650
```

```python
#Calculating and plotting the first derivative of the curve (Maximum rate of growth)
covid_agg.loc["China"].diff().plot() # Shows us the change of infection rate day by day 
covid_agg.loc["US"].diff().plot()
covid_agg.loc["Italy"].plot()
plt.legend()
```

```python
#Finding the Maximum Infection Rate For China, Italy, US
china_max = covid_agg.loc["China"].diff().max()
US_max = covid_agg.loc["US"].diff().max()
Italy_max = covid_agg.loc["Italy"].diff().max()

print("The maximum  single day infection rate for China : ", china_max)
print("The maximum single day infection rate for US is: ", US_max)
print("The maximum single day infection rate for Italy is: ", Italy_max)
```

```python
#Finding the maximum infection rate for all countries
#Create a new column for the max infection rates for each country   
countries = list(covid_agg.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(covid_agg.loc[c].diff().max())
covid_agg["max_infection_rate"] = max_infection_rates
```

```python
#Display updated dataset
covid_agg.head()
```

```python
#Create a new dataframe with the important columns

covid_data = pd.DataFrame(covid_agg["max_infection_rate"])
```

```python
covid_data.head()
```

```python
#importing World Happiness Report Dataset
world_happiness = pd.read_csv("worldwide_happiness.csv")
world_happiness.head()
```

```python
#Drop the useless columns
world_happiness.drop(["Overall rank", "Score", "Generosity", "Perceptions of corruption"], axis = 1, inplace = True)
```

```python
#Display updated dataset
world_happiness.head()
```

```python
#Changing the indicies of the dataframe

world_happiness.set_index("Country or region", inplace = True)
```

```python
world_happiness.head()
```

```python
#Join covid dataset with world happiness dataset
#check shape of both datasets
```

```python
covid_data.shape
```

```python
world_happiness.shape
```

```python
#Innerjoin the datasets because difference in shape
data = covid_data.join(world_happiness, how = "inner")
data.head()
```

```python
#Correlation Matrix 

data.corr() #positive correlation between infection rate and all the life factors of the dataset
```

```python
#Visualization with seaborn
```

```python
#Plotting GDP vs max infection rate

x = data["GDP per capita"]
y = data["max_infection_rate"]
sns.scatterplot(x, np.log(y)) 
```

```python
#regression plot
sns.regplot(x, np.log(y)) 
```

```python
#Plotting Social support vs max infection rate
x1 = data["Social support"]
sns.regplot(x1, np.log(y))
```

```python
#Plotting healthy life expectancy vs max infection rate
x2 = data["Healthy life expectancy"]
sns.regplot(x2, np.log(y))
```

```python
#Plotting Freedom to make life choices
x3 = data["Freedom to make life choices"]
sns.regplot(x3, np.log(y))
```
